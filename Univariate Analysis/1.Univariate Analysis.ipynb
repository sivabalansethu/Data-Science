# Measure the Central Tendency (Finding Mean, Median, Mode)

```python
dataset.describe()
```

```python
dataset["ssc_p"].mean()
```

```python
dataset["ssc_p"].median()
```

```python
dataset["ssc_p"].mode()[0]
```

```python
#In Table Column
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5Rule","Lesser","Greater","Min","Max"],columns=quan)
```

```python
descriptive
```

```python
descriptive["sl_no"]["99%"]=np.percentile(dataset["sl_no"],99)
```

```python
for descColName in quan:
    descriptive[descColName]["Mean"]=dataset[descColName].mean()
    descriptive[descColName]["Median"]=dataset[descColName].median()
    descriptive[descColName]["Mode"]=dataset[descColName].mode()[0]
    descriptive[descColName]["Q1:25%"]=dataset.describe()[descColName]["25%"]
    descriptive[descColName]["Q2:50%"]=dataset.describe()[descColName]["50%"]
    descriptive[descColName]["Q3:75%"]=dataset.describe()[descColName]["75%"]
    descriptive[descColName]["99%"]=np.percentile(dataset[descColName],99)
    descriptive[descColName]["Q4:100%"]=dataset.describe()[descColName]["max"]
    descriptive[descColName]["IQR"]=descriptive[descColName]["Q3:75%"]-descriptive[descColName]["Q1:25%"]
    descriptive[descColName]["1.5Rule"]=1.5*descriptive[descColName]["IQR"]
    descriptive[descColName]["Lesser"]=descriptive[descColName]["Q1:25%"]-descriptive[descColName]["1.5Rule"]
    descriptive[descColName]["Greater"]=descriptive[descColName]["Q3:75%"]-descriptive[descColName]["1.5Rule"]
    descriptive[descColName]["Min"]=dataset[descColName].min()
    descriptive[descColName]["Max"]=dataset[descColName].max()
```
