```python
quan=[]
qual=[]
for columnName in original_dataset.columns:
    #print(columnName)
    if(original_dataset[columnName].dtype=='O'):
        #print("qual")
        qual.append(columnName)
    else:
        #print("quan")
        quan.append(columnName)
```

# Univariate Analysis

```python
univariate_analysis=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%",
                               "Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater",
                                    "Min","Max","kurtosis","skew","Var","Std"],columns=quan)

lesserOutlier=[]
greaterOutlier=[]

for columnName in quan:
    univariate_analysis[columnName]["Mean"]=original_dataset[columnName].mean()
    univariate_analysis[columnName]["Median"]=original_dataset[columnName].median()
    univariate_analysis[columnName]["Mode"]=original_dataset[columnName].mode()[0]
    univariate_analysis[columnName]["Q1:25%"]=np.percentile(original_dataset[columnName], 25)
    univariate_analysis[columnName]["Q2:50%"]=np.percentile(original_dataset[columnName], 50)
    univariate_analysis[columnName]["Q3:75%"]=np.percentile(original_dataset[columnName], 75)
    univariate_analysis[columnName]["99%"]=np.percentile(original_dataset[columnName], 99)
    univariate_analysis[columnName]["Q4:100%"]=np.percentile(original_dataset[columnName], 100)
    univariate_analysis[columnName]["IQR"]=univariate_analysis[columnName]["Q3:75%"]-univariate_analysis[columnName]["Q1:25%"]
    univariate_analysis[columnName]["1.5rule"]=1.5*univariate_analysis[columnName]["IQR"]
    univariate_analysis[columnName]["Lesser"]=univariate_analysis[columnName]["Q1:25%"]-univariate_analysis[columnName]["1.5rule"]
    univariate_analysis[columnName]["Greater"]=univariate_analysis[columnName]["Q3:75%"]+univariate_analysis[columnName]["1.5rule"]
    univariate_analysis[columnName]["Min"]=original_dataset[columnName].min()
    univariate_analysis[columnName]["Max"]=original_dataset[columnName].max()
    univariate_analysis[columnName]["kurtosis"]=original_dataset[columnName].kurtosis()
    univariate_analysis[columnName]["skew"]=original_dataset[columnName].skew()
    univariate_analysis[columnName]["Var"]=original_dataset[columnName].var()
    univariate_analysis[columnName]["Std"]=original_dataset[columnName].std()
    
    #Find out Outlier Variable
    if(univariate_analysis[columnName]["Min"] < univariate_analysis[columnName]["Lesser"]):
        lesserOutlier.append(columnName)
    if(univariate_analysis[columnName]["Max"] > univariate_analysis[columnName]["Greater"]):
        greaterOutlier.append(columnName)

print(lesserOutlier)
print(greaterOutlier)

 
```

```python
#Replace outlier in Lesser and Greater
for columnName in lesserOutlier:
    original_dataset[columnName][original_dataset[columnName] < univariate_analysis[columnName]["Lesser"]]=univariate_analysis[columnName]["Lesser"]

for columnName in greaterOutlier:
    original_dataset[columnName][original_dataset[columnName] > univariate_analysis[columnName]["Greater"]]=univariate_analysis[columnName]["Greater"]
```
